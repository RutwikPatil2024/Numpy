#### Condition based selection

```python
arr=np.array([10,20,30,40,50,60,70,80,90])
arr
```

```python
bool_arr=[True,False,True,False,True,True,False,True,False]
arr[bool_arr]
```

```python
bool_arr=arr>50
arr[bool_arr]
```

```python
arr[arr<50]
```

```python
arr=np.arange(1,26).reshape(5,5)
arr
```

```python
arr=np.arange(1,26).reshape(5,5)
arr
```

```python
bool_arr=np.logical_and(arr%2==0,arr%10==0)
bool_arr
```

```python
arr[bool_arr]
```

```python
arr[(arr%2==0) & (arr%5==0)]
```

```python
bool_arr=np.logical_or(arr%2==0,arr%3==0)
bool_arr
```

```python
arr[bool_arr]
```

```python
arr[(arr%2==0) | (arr%3==0)]
```
