```python
# positional argument and keyword argument
def func(a,b,/,c,d):
    print("value of a : ",a)
    print("value of b : ",b)
    print("value of c : ",c)
    print("value of d : ",d)
func(10,20,30,40)
```

```python
def func(a,b,/,c,d):
    print("value of a : ",a)
    print("value of b : ",b)
    print("value of c : ",c)
    print("value of d : ",d)
func(a=10,20,30,40)
# before / all must be passed as positional argument
```

```python
def func(a,b,/,c,d):
    print("value of a : ",a)
    print("value of b : ",b)
    print("value of c : ",c)
    print("value of d : ",d)
func(10,20,c=30,d=40)
# After / we can pass as positional argument or keyword argument
```

```python
def func(a,b,*,c,d):
    print("value of a : ",a)
    print("value of b : ",b)
    print("value of c : ",c)
    print("value of d : ",d)
func(10,20,30,40)
# before * pass as keyword argument 
```
